```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
start_date = "2025-01-01"
end_date = "2026-01-01"
df = yf.download(["LT.NS" , "INFY" , "JPM"], start=start_date, end=end_date)
```

```python
daily_returns_matrix = pd.DataFrame({
    "LT.NS": df["Close"]["LT.NS"].pct_change(),
    "INFY": df["Close"]["INFY"].pct_change(),
    "JPM": df["Close"]["JPM"].pct_change()
})
daily_returns_matrix
```

```python
daily_returns_matrix.dropna(inplace=True)
daily_returns_matrix
```

```python
avg_daily_return_LTNS = daily_returns_matrix["LT.NS"].mean()
avg_daily_return_INFY = daily_returns_matrix['INFY'].mean()
avg_daily_return_JPM = daily_returns_matrix['JPM'].mean()
print(f"Average daily return for LT.NS: {avg_daily_return_LTNS}")
print(f"Average daily return for INFY: {avg_daily_return_INFY}")
print(f"Average daily return for JPM: {avg_daily_return_JPM}")
```

```python
risk_free_rate = 0.0001
```

```python
sharpe_LTNS = (avg_daily_return_LTNS - risk_free_rate)/daily_returns_matrix["LT.NS"].std()
sharpe_INFY = (avg_daily_return_INFY - risk_free_rate)/daily_returns_matrix['INFY'].std()
sharpe_JPM = (avg_daily_return_JPM - risk_free_rate)/daily_returns_matrix['JPM'].std()
print(f"Sharpe ratio for LT.NS: {sharpe_LTNS}")
print(f"Sharpe ratio for INFY: {sharpe_INFY}")
print(f"Sharpe ratio for JPM: {sharpe_JPM}")
```

```python
volatility_LTNS = daily_returns_matrix["LT.NS"].std()
volatility_INFY = daily_returns_matrix['INFY'].std()
volatility_JPM = daily_returns_matrix['JPM'].std()
print(f"Volatility for LT.NS: {volatility_LTNS}")
print(f"Volatility for INFY: {volatility_INFY}")
print(f"Volatility for JPM: {volatility_JPM}")
```

```python
correlation_matrix = daily_returns_matrix.corr()
correlation_matrix
```

```python
covariance_matrix = daily_returns_matrix.cov()
covariance_matrix
```

```python
rolling_mean_LTNS = daily_returns_matrix["LT.NS"].rolling(window=30).mean()
rolling_mean_INFY = daily_returns_matrix['INFY'].rolling(window=30).mean()
rolling_mean_JPM = daily_returns_matrix['JPM'].rolling(window=30).mean()
print(f"Rolling mean for LT.NS: {rolling_mean_LTNS}")
print(f"Rolling mean for INFY: {rolling_mean_INFY}")
print(f"Rolling mean for JPM: {rolling_mean_JPM}")
```

```python
plt.plot(daily_returns_matrix["LT.NS"], label="LT.NS")
plt.plot(daily_returns_matrix['INFY'], label='INFY')
plt.plot(daily_returns_matrix['JPM'], label='JPM')
plt.legend()
plt.title('Daily Returns of Stocks')
plt.xlabel('Date')
plt.ylabel('Daily Return')
plt.show()
```

```python
plt.plot(rolling_mean_LTNS, label="LT.NS")
plt.plot(rolling_mean_INFY, label='INFY')
plt.plot(rolling_mean_JPM, label='JPM')
plt.legend()
plt.title('Rolling Mean of Daily Returns of Stocks')
plt.xlabel('Date')
plt.ylabel('Rolling Mean')
plt.show()
```

```python
stocks = ['LT.NS', 'INFY', 'JPM']
sharpes = [sharpe_LTNS, sharpe_INFY, sharpe_JPM]
plt.bar(stocks, sharpes,color = ['red','blue','green'])
plt.legend()
plt.title('Sharpe Ratios of Stocks')
plt.xlabel('Stock')
plt.ylabel('Sharpe Ratio')
plt.show()
```

Off the three stocks, JP Morgan and Chase (ticker JPM) showed the best risk-adjusted return with a significantly high sharpe ratio and the corresponding high average daily returns and low volatility

Indian stocks (L&T, Infosys) reveal an appalling correlation pattern with the USA stock (JPMC)
L&T shows low correlation with with both JPMC as well as Infosys whereas Infosys and JPMC have a relatively high correlation

In the market dip of 2025, banking sector proved to be resilient compared to infrastructure and IT sector with low volatility and high risk-adjusted returns
