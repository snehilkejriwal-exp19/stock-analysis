**CAPM AND Beta Estimation for Public Companies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
ticker = ['^GSPC' , 'BX' , 'ACN' , 'CTSH'] #S&P 500, BLACKSTONE, ACCENTURE, COGNIZANT
start_date = "2025-01-01"
end_date = "2026-01-01"
df = yf.download(ticker,start_date,end_date)
df
```

```python
#Beta = Covariance(stock returns, market returns) / Variance(market returns)
```

```python
returns = df['Close'].pct_change(fill_method=None).dropna()
market_returns = returns['^GSPC']
stock_returns = returns[['BX', 'ACN', 'CTSH']]
```

```python
beta_bx = np.cov(stock_returns['BX'], market_returns)[0, 1] / np.var(market_returns)
beta_acn = np.cov(stock_returns['ACN'], market_returns)[0 ,1] / np.var(market_returns)
beta_ctsh = np.cov(stock_returns['CTSH'] , market_returns)[0,1] /np.var(market_returns)
print(f"The beta for Blackstone is {beta_bx}")
print(f"The beta for Accenture is {beta_acn}")
print(f"The beta for Cognizant is {beta_ctsh}")
```

```python
risk_free_rate = 0.045
```

```python
market_annual_return = (
    df['Close']['^GSPC'].iloc[-1]
    - df['Close']['^GSPC'].iloc[0]
) / df['Close']['^GSPC'].iloc[0]
```

```python
expected_return_bx = risk_free_rate + beta_bx * (market_annual_return - risk_free_rate)
expected_return_acn = risk_free_rate + beta_acn * (market_annual_return - risk_free_rate)
expected_return_ctsh = risk_free_rate + beta_ctsh * (market_annual_return - risk_free_rate)
```

```python
first_price_bx = df['Close']['BX'].iloc[0]
last_price_bx = df['Close']['BX'].iloc[-1]
actual_annual_return_bx = (last_price_bx - first_price_bx) / first_price_bx
```

```python
first_price_acn = df['Close']['ACN'].iloc[0]
last_price_acn = df['Close']['ACN'].iloc[-1]
actual_annual_return_acn = (last_price_acn - first_price_acn) / first_price_acn
```

```python
first_price_ctsh = df['Close']['CTSH'].iloc[0]
last_price_ctsh = df['Close']['CTSH'].iloc[-1]
actual_annual_return_ctsh = (last_price_ctsh - first_price_ctsh) / first_price_ctsh
```

```python
alpha_bx = actual_annual_return_bx - expected_return_bx
alpha_acn = actual_annual_return_acn - expected_return_acn
alpha_ctsh = actual_annual_return_ctsh - expected_return_ctsh
```

```python
final = pd.DataFrame({
    "stock": ["BX", "ACN", "CTSH"],
    "start_date": [start_date, start_date, start_date],
    "end_date": [end_date, end_date, end_date],
    "beta": [beta_bx, beta_acn, beta_ctsh],
    "alpha": [alpha_bx, alpha_acn, alpha_ctsh],
    "annual_return": [
        actual_annual_return_bx,
        actual_annual_return_acn,
        actual_annual_return_ctsh
    ],
    "expected_return": [
        expected_return_bx,
        expected_return_acn,
        expected_return_ctsh
    ]
})

final[['beta']] = final[['beta']].round(3)
final[['alpha','annual_return','expected_return']] = final[['alpha','annual_return','expected_return']].applymap(lambda x: f"{x:.2%}")

final
```

```python
#Blackstone emerged with a beta of 1.569 which means that the stock is more volatile than the market while Accenture and Cognizant showed a beta of 0.790 and 0.865 respectively meaning that the stock is less volatile than the market
#Cognizant showed the least negative alpha (-4.62%) meaning the difference between actual return and expected return is much lesser compard to blackstone and accenture making it a more reliable stock for an investor
#Analyzing the table we see that all the stocks underperformed with respect to CAPM expectations showing high volatility in the market paired with uncertainty.
```
