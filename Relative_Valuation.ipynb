```python
import numpy as np
import pandas as pd
```

```python
#VISA
current_share_price_visa = 323.57
eps_visa = 10.74
book_value_per_share_visa = 17.40
EBITDA_visa = 25.214
market_cap_visa = 552.47
```

```python
#ICICI
current_share_price_icici = 26.17
eps_icici = 1.68
book_value_per_share_icici = 6.01
EBITDA_icici = 0.00
market_cap_icici = 92.04
```

```python
#SNOWFLAKE
current_share_price_snowflake = 238.26
eps_snowflake = 3.52
book_value_per_share_snowflake = 18.32
EBITDA_snowflake = -1.088
market_cap_snowflake = 84.34
```

```python
PE_visa = current_share_price_visa / eps_visa
PE_icici = current_share_price_icici/ eps_icici
PE_snowflake = current_share_price_snowflake/ eps_snowflake
print(f"The P/E ratio for Visa is {PE_visa} dollars")
print(f"The P/E ratio for ICICI Bank is {PE_icici} dollars")
print(f"The P/E ratio for Snowflake is {PE_snowflake} dollars")
```

```python
PB_visa = current_share_price_visa / book_value_per_share_visa
PB_icici = current_share_price_icici/ book_value_per_share_icici
PB_snowflake = current_share_price_snowflake/ book_value_per_share_snowflake
print(f"The P/B ratio for Visa is {PB_visa} dollars")
print(f"The P/B ratio for ICICI Bank is {PB_icici} dollars")
print(f"The P/B ratio for Snowflake is {PB_snowflake} dollars")
```

```python
ev_by_ebidta_visa = market_cap_visa / EBITDA_visa
ev_by_ebidta_snowflake = market_cap_snowflake / EBITDA_snowflake
print(f"The EV/EBITDA ratio for Visa is {ev_by_ebidta_visa} dollars")
print(f"The EV/EBITDA ratio for Snowflake is {ev_by_ebidta_snowflake} dollars")
print("EBITDA is not a useful metric for evaluating a bank, hence EBITDA of ICICI Bank is 0")
```

```python
df = pd.DataFrame({
    "Company": ["Visa", "ICICI", "Snowflake"],
    "PE": [PE_visa, PE_icici, PE_snowflake],
    "PB": [PB_visa, PB_icici, PB_snowflake],
    "EV/EBITDA": [ev_by_ebidta_visa, 0, ev_by_ebidta_snowflake]
}
    )
df = df.round(2)
df
```
